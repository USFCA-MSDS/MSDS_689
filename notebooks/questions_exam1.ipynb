(7). Tree Properties: What is the height of a tree? How is it different from the depth of a node?

    Solution: The height of a tree is the maximum distance from the root node to any leaf node. The depth of a node is the distance from the root node to that particular node.





```python
# (8) Given the following Python code:

def dfs(graph, start, visited=None):
    if visited is None:
        visited = set()
    visited.add(start)
    for neighbor in graph[start]:
        if neighbor not in visited:
            dfs(graph, neighbor, visited)

def is_connected(graph):
    start_vertex = next(iter(graph.keys()))  # Start DFS from an arbitrary vertex
    visited = set()
    dfs(graph, start_vertex, visited)
    return len(visited) == len(graph)

# Example usage:
graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D'],
    'C': ['A', 'D'],
    'D': ['B', 'C'],
    'E': ['F'],
    'F': ['E']
}

print("Is the graph connected?", is_connected(graph))
```
